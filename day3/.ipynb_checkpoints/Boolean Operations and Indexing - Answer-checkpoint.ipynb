## Excercise 1

```python
def find_div(arr, val):
    bool_ind = (arr % val == 0)
    row, col = np.where(bool_ind)
    
    for r, c in zip(row, col):
        print(arr[r,c], "| location: row ", r, "col ", c)
```

```python
a = np.random.randint(0,20,(5,5))
a
```

```python
find_div(a, 10)
```

## Exercise 2

```python
signal_1 = np.array(([0] * 20 + [1] * 20) * 10)
signal_2 = np.array(([0] * 15 + [1] * 20 + [0] * 5) * 10)
timing = np.linspace(0, 60, signal_1.size)

fig, ax = plt.subplots(figsize = (20,5))
ax.plot(timing, signal_1, timing, signal_2)
plt.xlabel('Time (sec)')
plt.ylabel('Amplitude')
plt.show()
```

### Part 1

```python
signal_intersect = np.zeros(signal_1.shape)
signal_intersect[(signal_1 == 1) & (signal_2 == 1)] = 1

fig, ax = plt.subplots(figsize=(20,5))
ax.plot(timing, signal_1, timing, signal_2)
ax.plot(timing, signal_intersect, linewidth=5, color='g')
plt.xlabel('Time (sec)')
plt.ylabel('Amplitude')
plt.show()
```

### Part 2

```python
signal_intersect_diff = np.zeros(signal_intersect.shape)
signal_intersect_diff[1:] = np.diff(signal_intersect)

fig, ax = plt.subplots(figsize=(20,5))
ax.plot(timing, signal_intersect_diff, linewidth=5, color='g')
plt.xlabel('Time (sec)')
plt.ylabel('Amplitude')
plt.show()
```

### Part 3

```python
starting_times = timing[signal_intersect_diff == 1]
ending_times = timing[signal_intersect_diff == -1]

period = ending_times - starting_times

fig, ax = plt.subplots()
ax.hist(period)
plt.xlabel('Time (sec)')
plt.ylabel('Frequency')
plt.show()
```
